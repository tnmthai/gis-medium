```python
mtcars.shape
```

Barplot


```python
res = sns.barplot(x=mtcars['cyl'], y=mtcars['carb'])
plt.show()
```

```python
res = sns.barplot(x='cyl', y='carb', data=mtcars)
plt.show()
```

```python
res = sns.barplot(x='cyl', y='carb', data=mtcars, color='red')
plt.show()
```

```python
res = sns.barplot(x=mtcars['gear'], y=mtcars['hp'], hue=[], palette='rocket')
plt.show()
```
